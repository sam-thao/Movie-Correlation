```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create

# Read in the data

df = pd.read_csv(r'C:\Users\Owner\Desktop\Movie Industry\movies.csv')
```

```python
# lets look at the data

df.head() # method returns a specified number of rows, string from the top.
```

```python
# lets see if there's any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    df = df.dropna() # drop any data that is missing
```

```python
# Data types for our columns

df.dtypes # is an attribute of arrays that represents the data type of the elements in the array
```

```python
# Change data type of columns

df['budget'] = df['budget'].astype('int64') 

df['gross'] = df['gross'].astype('int64') 
```

```python
df.head()
```

```python
# Create correct Year column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(str)

df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False) # Sorting by gross column, descending
```

```python
pd.set_option('display.max_rows', None) #displaying max rows
```

```python
df.head()
```

```python
# Scatter plot with budget vs gross revenue

plt.scatter(x=df['budget'], y=df['gross']) # method used for creating scatter plots

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show() #function is used to display the plot that you have created 
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()

# Budget to gross is .74
# Votes to Gross .61
```
